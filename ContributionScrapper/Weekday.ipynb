```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import glob, os
from datetime import datetime
import seaborn as sns
```

```python
def convert_to_date(date):
    return datetime(year=int(date[0:4]),month=int(date[5:7]),day=int(date[8:10]))

l = []
c = 1

os.chdir("./GithubContributions")
for file in glob.glob("*.xlsx"):
    weekday = 0
    weekend = 0
    data = pd.read_excel(file,sheetname='sheet1')
    data = data.drop_duplicates()
    dates = list(data['Date'])
    contributions = list(data['XContributions'])
    for i in range(len(dates)):
        if(convert_to_date(dates[i]).weekday()<5):
            weekday+=contributions[i]
        else:
            weekend+=contributions[i]
    ratio = float(weekend)/float(weekend+weekday)
    l.append(ratio)
    print file + ' ' + str(c)
    c = c+1
```

```python
plt.figure(figsize=(10,5))
p1 = sns.distplot(l,bins=20,kde=False)
p1.set(xlabel='Ratio of commits on weekends to all days', ylabel='No. of users')
plt.show()
```

```python
np.mean(l)
# ratio of weekdays to all days = 2/7 = 0.28
```
